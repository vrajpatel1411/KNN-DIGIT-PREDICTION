```python
import numpy as np
from matplotlib import pyplot


def function_of_Beta(Beta):
    return Beta**2.0


def derivative_of_Beta(Beta):
    
    return Beta*2.0

def gradient_descent(learning_rate):
    solutions, scores = list(), list()
    Beta = 4.5
    solutions.append(Beta)
    scores.append(function_of_Beta(Beta)) 
    for k in range(1000):
        Beta = Beta -(learning_rate * derivative_of_Beta(Beta))
        solutions.append(Beta)
        scores.append(function_of_Beta(Beta))
    return solutions,scores


learning_rate=0.01  
solutions,scores=gradient_descent(learning_rate)
bounds=np.arange(-5.0,6.0)
results = function_of_Beta(bounds)
pyplot.plot(bounds,results)
pyplot.plot(solutions, scores, '.-',color='red')
pyplot.xlabel("range of beta")
pyplot.ylabel("function of beta")
title="Graph where learning rate is "+str(learning_rate)
pyplot.title(title)
pyplot.show()
```
